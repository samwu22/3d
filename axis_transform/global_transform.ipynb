```python
# Camera golbal
import numpy as np
from sympy import *
# ----------------cam golbal---------------
x1=-25.554613; y1=89.03798; z1=1061.2738
x2=6.1871133; y2=92.79199;z2=1067.9972
x3=22.497286; y3=202.63237; z3=965.96857
x4=94.445465; y4=195.76158; z4=981.11053

# reference box axis
# y
# |
# ------x
xb1=5 ;yb1=181;zb1=100
xb2=38 ;yb2=177;zb2=100
xb3=34 ;yb3=67;zb3=0
xb4=107 ;yb4=73;zb4=0
# coordinate = x, y = 1017, 988  #U  (-25.554613, 89.03798, 1061.2738, 103, 175, 210)
# coordinate = x, y = 1087, 995  #绿 (6.1871133, 92.79199, 1067.9972, 142, 203, 255)
# coordinate = x, y = 1128, 1283  #日 22.497286, 202.63237, 965.96857, 106, 94, 79)
# coordinate = x, y = 1299, 1259  #内  (94.445465, 195.76158, 981.11053, 113, 86, 74)
```

```python
# from https://blog.csdn.net/wangxiaokun671903/article/details/37966891
# 二元一次方程
# =Symbol('')
r1=Symbol('r1')
r2=Symbol('r2')
r3=Symbol('r3')
r4=Symbol('r4')
r5=Symbol('r5')
r6=Symbol('r6')
r7=Symbol('r7')
r8=Symbol('r8')
r9=Symbol('r9')
t1=Symbol('t1')
t2=Symbol('t2')
t3=Symbol('t3')

cam_2_ref=[
    xb1-(r1*x1+r2*y1+r3*z1+t1),
    yb1-(r4*x1+r5*y1+r6*z1+t2),
    zb1-(r7*x1+r8*y1+r9*z1+t3),
    
    xb2-(r1*x2+r2*y2+r3*z2+t1),
    yb2-(r4*x2+r5*y2+r6*z2+t2),
    zb2-(r7*x2+r8*y2+r9*z2+t3),
    
    xb3-(r1*x3+r2*y3+r3*z3+t1),
    yb3-(r4*x3+r5*y3+r6*z3+t2),
    zb3-(r7*x3+r8*y3+r9*z3+t3),
    
    xb4-(r1*x4+r2*y4+r3*z4+t1),
    yb4-(r4*x4+r5*y4+r6*z4+t2),
    zb4-(r7*x4+r8*y4+r9*z4+t3),
]

ref_2_cam=[
    x1-(r1*xb1+r2*yb1+r3*zb1+t1),
    y1-(r4*xb1+r5*yb1+r6*zb1+t2),
    z1-(r7*xb1+r8*yb1+r9*zb1+t3),
    
    x2-(r1*xb2+r2*yb2+r3*zb2+t1),
    y2-(r4*xb2+r5*yb2+r6*zb2+t2),
    z2-(r7*xb2+r8*yb2+r9*zb2+t3),
    
    x3-(r1*xb3+r2*yb3+r3*zb3+t1),
    y3-(r4*xb3+r5*yb3+r6*zb3+t2),
    z3-(r7*xb3+r8*yb3+r9*zb3+t3),
    
    x4-(r1*xb4+r2*yb4+r3*zb4+t1),
    y4-(r4*xb4+r5*yb4+r6*zb4+t2),
    z4-(r7*xb4+r8*yb4+r9*zb4+t3),
]

us=[r1,r2,r3,r4,r5,r6,r7,r8,r9,t1,t2,t3]
solved_cam_2_ref=solve(cam_2_ref, us)
solved_ref_2_cam=solve(ref_2_cam, us)

RT=[];R=[];T=[]
for k,v in solved_ref_2_cam.items():
    if k == t1 or k==t2 or k==t3:
        T.append(v)
    else:
        R.append(v)
    RT.append(v)
R=np.array(R);R=R.reshape(3,3)
T=np.array(T);T=T.reshape(3,1)
RT=np.array(RT)
RT=RT.reshape(3,4)

RT2=[];R2=[];T2=[]
for k,v in solved_cam_2_ref.items():
    if k == t1 or k==t2 or k==t3:
        T2.append(v)
    else:
        R2.append(v)
    RT2.append(v)
R2=np.array(R2);R2=R2.reshape(3,3)
T2=np.array(T2);T2=T2.reshape(3,1)
RT2=np.array(RT2)
RT2=RT2.reshape(3,4)

# test point in ref
ref_x=5
ref_y=181
ref_z=100
temp=np.array([ref_x,ref_y,ref_z,1]).reshape(4,1)
ans=np.dot(RT,temp)
print(ans)


# test point in cam
cam_x=-25.554613; cam_y=80.03798; cam_z=1061.2738
temp2=np.array([cam_x,cam_y,cam_z,1]).reshape(4,1)
ans2=np.dot(RT2,temp2)
print(ans2)
```
